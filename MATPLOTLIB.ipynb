# Stylish Graph

```python
from matplotlib import style 
style.use('ggplot')
```

```python
x=[1,2,6]
y=[9,3,4]
```

```python
plt.plot(x,y,'g',label='Line one',linewidth=5)
plt.show()
```

```python
x1=[9,2,3]
y1=[5,1,7]
x=[1,2,6]
y=[9,3,4]
```

```python
plt.plot(x1,y1,'r',label='Line one',linewidth=5)
plt.plot(x,y,'c',label='Line two',linewidth=5)
plt.show()
```
